```python
#standard imports
import pandas as pd
import numpy as np
np.random.seed(123)#the random numbers will be the same for another user or if rerun later
```

```python
#randomnly generate scores for student list
students=['Sally','Jane','Suzie','Billy','Ada','John','Thomas','Marie','Albert','Richard','Issac','Alan']
math_grades=np.random.randint(low=60,high=100,size=len(students))
english_grades=np.random.randint(low=60,high=100,size=len(students))
reading_grades=np.random.randint(low=60,high=100,size=len(students))

df=pd.DataFrame({'name':students,
                'math':math_grades,
                'english':english_grades,
                'reading':reading_grades})
print(type(df))
print(df)
```

Series are back! Each column of the dataframe is actually a series and you can use all the series methods on it. To access individual columns use the dataframe.column_name or the bracket notation. Bracket notation must be used if there a spaces/weird characters in column names

```python
#The head and tail can be used as they were used in series. New is sample that brings back a random
#sample of size n
print(df.head())
#print(df.tail())
#print(df.sample())
```

```python
fail=df.math<70#boolean series that can be used to subset a dataframe; can be saved in variable or used
df[fail]
```

Pay attention to default behavior of .drop and .rename. They will create **new** dataframe. If that is not what you want change inplace=True. Very important with larger dataframes

```python
df.drop(columns=['english','reading'],inplace=False)
```

```python
df.rename(columns={'math':'maths'})
```

```python
#Method chaining is very common in pandas. Be careful if you try to use inplace=True as that will not return a dataframe
df.drop(columns=['english','reading']).rename(columns={'math':'maths'})
```

```python
df.math<70
```

```python
df['failing_math']=df.math<70
df.assign(passing_reading=df.reading>=70)
```

```python
#notice the column with .assign did not show up since by default it creates a new dataframe
df
```

```python
# we can also sort by a particular column or columns(make them a list)
df.sort_values(by=['english','math'],ascending=False)
```

```python
#Revisiting chaining. Again don't use this with larger dataframes
df[df.math>50].sort_values(by='math').tail(5).name
```
