Always start with your imports
First we will cover line plots using the the .plot() command

```python
import matplotlib.pyplot as plt
import math
import random
%matplotlib inline 

x = list(range(150))#list of value from 0 to 149
plt.plot(x,c='red')#use the 'c' arguement to specify color
plt.xlim(-20, 200)#change the limits of graph
plt.ylim(-5, 160)#change the limits of graph
plt.show()
```

plotting two series on the same graph

```python
from random import randint

# Some random data
x1 = [randint(-5, 25) + n for n in range(150)]
x2 = [x + randint(-6, 6) for x in x1]

# we can use the alpha kwarg to define how transparent the color is
plt.plot(x1, c='green', alpha=0.6)
plt.plot(x2, c='red', alpha=0.4)
plt.show()
```

Scatter plot along with labeling axes and titling graph 

```python
plt.scatter(range(len(x1)), x1, s=30, c='red')
plt.scatter(range(len(x2)), x2, s=10, c='green')
#plt.title('A couple of random series')#title can be added,must be before the show method
#plt.xlabel('$x$')#use xlabel to label your x series
#plt.ylabel('$y$')#use ylabel to label your y series
plt.show()
```

```python
x = [1, 2, 3.14, 4]
y = [n ** 2 for n in x]
plt.scatter(x, y)
plt.plot(x,y,ls=':')#specifying the style for the line
plt.xticks([1, 2, 3.14, 4], ['one', 'two', '$\pi$', 'four'])#custom tick marks on x-axis
plt.yticks(rotation=45)#rotation arguement can be set to change orientation of axis labels
plt.show()
```

Having all this control allows labels to not overlap and be more readable

```python

weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'] * 2#remember that multiplying a string makes n copies of values

x = range(14)
y = [random.randint(1, 10) for _ in x]
plt.plot(x, y)
plt.xticks(x, weekdays, rotation=60)
plt.show()
```

You can also add text boxes and annotations to make things stand out or to draw user attention
to certain features

```python
x = list(range(-4, 4))
x.append(3.14159)
y = [math.sin(n) for n in x]

plt.scatter(x, y)
plt.plot(x, y, ls=':')

plt.text(0.25, 0, '(0, 0)', fontsize=10, color='black')
plt.text(1.2, 0.1, '($\pi$, sin($\pi$))', fontsize=14, color='firebrick')
plt.annotate('Origin', xy=(0, 0), xytext=(2, 1),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

A legend is also availabile in matplotlib
and can be customized using the legend method.
Plots can also be saved and this can be accomplished in plot window
or programmatically

```python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # (width, height)

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')

plt.legend(loc='upper right')
plt.title('Approximate sin and cosine curves')
plt.xlabel('$x$')
plt.savefig('trig-figure')
plt.show()
```

```python
Subplots are useful when you want to display related
graphs next to each other for comparison.
```

```python
# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(121)
plt.plot(x, y)
plt.title('x ~ y')

# the second subplot
plt.subplot(122)
plt.plot(x, z)
plt.title('x ~ z')

plt.suptitle('Subplots Demo')#overall title

plt.show()
```
